**2.- Modificar la función dqr() para que en el reporte se agregue una columna con las categorías de las columnas (valores unicos) que resulten ser categóricas (si las categorias > 10, no agregar esas categorias)**


```python
def dqr(data):
    cols = pd.DataFrame(list(data.columns.values), columns=['Nombres'], index=list(data.columns.values))
    cols['Tipo_Dato'] = data.dtypes

    cols['Is_categorical'] = ~data.apply(lambda s: pd.api.types.is_numeric_dtype(s))

    def obtener_categorias(col_name):
        if cols.loc[col_name, 'Is_categorical']:
            valores_unicos = data[col_name].unique()
            if len(valores_unicos) <= 10:
                return list(valores_unicos)
        return np.nan

    cols['Categorias'] = cols.index.map(obtener_categorias)

    return cols
reporte = dqr(df_countries)
reporte.head()
```

**3.- Aplicar el análisis estadístico de las columnas creadas (max, min, mean, std) solo para las columnas que sean numéricas ('Is_categorical' == False)**


```python
def dqr(data):
    cols = pd.DataFrame(list(data.columns.values), columns=['Nombres'], index=list(data.columns.values))
    cols['Tipo_Dato'] = data.dtypes

    cols['Is_categorical'] = ~data.apply(lambda s: pd.api.types.is_numeric_dtype(s))

    def obtener_categorias(col_name):
        if cols.loc[col_name, 'Is_categorical']:
            valores = data[col_name].unique()
            if len(valores) <= 10:
                return list(valores)
        return np.nan
    cols['Categorias'] = cols.index.map(obtener_categorias)

    cols['Missing_Values'] = data.isnull().sum()

    for col in data.columns:
        if not cols.loc[col, 'Is_categorical']:
            cols.loc[col, 'Min'] = data[col].min()
            cols.loc[col, 'Max'] = data[col].max()
            cols.loc[col, 'Mean'] = data[col].mean()
            cols.loc[col, 'Std'] = data[col].std()
        else:
            cols.loc[col, 'Min'] = np.nan
            cols.loc[col, 'Max'] = np.nan
            cols.loc[col, 'Mean'] = np.nan
            cols.loc[col, 'Std'] = np.nan
    return cols
reporte_final = dqr(df_countries)
reporte_final
```
